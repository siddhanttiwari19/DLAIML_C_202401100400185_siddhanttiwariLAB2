```python
X=(0.68,0.95,0.87,0.76)
w=[0.4,0.3,0.5,0.3] # Changed to a list to allow modification
b=-0.2
Y=1
n=0.01

z=X[0]*w[0]+X[1]*w[1]+X[2]*w[2]+X[3]*w[3]+b

if(z>=0):
  y_pred=1
else:
  y_pred=0

error=Y-y_pred # Calculate error outside the if-else block

if error != 0: # Update weights and bias only if there's an error
  w[0]=w[0]+(n*error*X[0])
  w[1]=w[1]+(n*error*X[1])
  w[2]=w[2]+(n*error*X[2])
  w[3]=w[3]+(n*error*X[3])
  b=b+(n*error)

# Recalculate z with potentially updated weights/bias
z=X[0]*w[0]+X[1]*w[1]+X[2]*w[2]+X[3]*w[3]+b

print(z) # Print the final z value
```
